```python
# Drop 'Resolution' as it's too messy
df.drop(['Resolution'], axis=1, inplace=True)

# Convert time columns to datetime
df['First Response Time'] = pd.to_datetime(df['First Response Time'], errors='coerce')
df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'], errors='coerce')

# Create a duration in minutes
df['Resolution Time (mins)'] = (df['Time to Resolution'] - df['First Response Time']).dt.total_seconds() / 60

# Drop original time columns
df.drop(['First Response Time', 'Time to Resolution'], axis=1, inplace=True)

# Fill missing satisfaction rating with median
df['Customer Satisfaction Rating'].fillna(df['Customer Satisfaction Rating'].median(), inplace=True)
```

```python
# Convert satisfaction to categories
df['Satisfaction_Level'] = pd.cut(df['Customer Satisfaction Rating'],
                                  bins=[0, 2, 3, 5],
                                  labels=['Low', 'Medium', 'High'])
df.drop('Customer Satisfaction Rating', axis=1, inplace=True)
```

```python
categorical_cols = ['Customer Gender', 'Product Purchased', 'Ticket Type',
                    'Ticket Subject', 'Ticket Status', 'Ticket Priority', 'Ticket Channel']

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.drop(['Ticket ID', 'Customer Name', 'Customer Email', 'Date of Purchase', 'Ticket Description'], axis=1, inplace=True)
df
```
